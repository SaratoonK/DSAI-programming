```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

boston = load_boston()
X = boston.data
y = boston.target
m = X.shape[0]  
n = X.shape[1]

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
intercept = np.ones((X_train.shape[0], 1))

X_train = np.concatenate((intercept, X_train), axis=1)

intercept = np.ones((X_test.shape[0], 1))

X_test = np.concatenate((intercept, X_test), axis=1)
print("x_train.shape is",X_train.shape)
print("y_train.shape is",y_train.shape)


alpha = 0.001
max_iter = 1000
loss_old = 10000
tol = 0.0001
iter_stop = 0
method = "minibatch"
used_num = []

def h_theta(X, theta):
    return X @ theta

def mse(yhat, y_i):
    return ((yhat - y_i)**2).sum() / yhat.shape[0]

def gradient(X_i, error):
    return X_i.T @ error


for i in range(max_iter):
    if method == "stochastic":
        i = np.random.randint(X_train.shape[0])
        if i not in used_num:
            used_num.append(i)
        else:
            i = np.random.randint(X_train.shape[0])
        
        X_i = X_train[i]
        
        X_train = X_i.reshape(1,-1)
        
        theta = np.zeros(X_train.shape[1])
        
        y_i = y_train.reshape(-1,1)
        
        y_i = y_i[i]
        
    elif method == "minibatch":
        
        r = np.random.randint(X_train.shape[0])
        #print(r)
        batch = int(0.3*X_train.shape[0])
        #print("batch is", batch)
        
        X_i = X_train[r:r+batch]
        
        #print("X_train shape is ", X_i.shape)
        
        #print("X_train shape is ", X_train.shape)
        theta = np.zeros(X_i.shape[1])
        
        #print("theta shape is ", theta.shape)
        
        y_i = y_train[r:r+batch]
        #print("y_i is ", y_i.shape)
        
        
    yhat = h_theta(X_i, theta)
    #print("yhat is ", yhat.shape)
    error = yhat - y_i
    #print("error shape is ",error.shape)
    
    loss_new = mse(yhat,y_i)
    if np.abs(loss_new-loss_old) < tol:
        break
 
    grad = gradient(X_i, error)

    theta = theta - alpha * grad
    
    
yhat = h_theta(X_test, theta)

mse = mse(yhat,y_test)
print(mse)


```
