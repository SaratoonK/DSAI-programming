```python
# assignment do linear probing and delete
# m should be larger than input
class HashTable:
    
    def __init__(self, m):
        self.m = m
        self.hashtable = self.create_hash_table()
    
    def create_hash_table(self):
        return [ [] for _ in range(self.m) ]

    def _prehash(self,key):
        if (type(key) == str):
            key = hash(key)  #returns a number for you
            
        if ((type(key) == int) | (type(key) == float)):
            if (key < 0):
                key = hash(float(key)) * -1  #first convert to float, then hash it
        
        return key
    def _hash(self,key):
         #get the position using division method
        index = key % self.m
        bucket = self.hashtable[index]
        # print("bucket number: ",index)
        return bucket
    def insert(self, key, val):
    
        key = self._prehash(key)
        # print("After prehash:", key)
        # assert (key > 0) & (type(key) == int)
        bucket = self._hash(key)
        # print("bucket",bucket)
        
                
        found, pos_dup, _ , bucket, _,missing = self.search(key)
    #if the key duplicates, only update the value
        if(found) or missing:
            # print(pos_dup)
            bucket[pos_dup] = (key, val)
        else: #if the key does not exist, append and #if something is there already, append
            # new_key = key
            # print("this is",bucket)
            # while len(bucket) != 0:
            #     new_key = new_key + 1
            #     bucket = self._hash(new_key)
            #     print(len(bucket))
            bucket.append((key, val))
            
        print(self.hashtable)
        
    def search(self, key,status = False):
        #if you finish this, 
        key = self._prehash(key)
        # print("Key after prehash:", key)
        #perform the division method
        bucket = self._hash(key)     #get the position of the hashtable
        
        new_key = key
     

        found  = False
        missing = False
        pos_dup = -999
        answer = -9999
        flag = -999
        while len(bucket) != 0:
        #loop the bucket index
            for i, (bkey, bval) in enumerate(bucket):
                # print(i,bkey,bval,bucket,new_key)
                if bkey == key:
                    found = True
                    pos_dup = i
                    answer = bval
                    break
                elif bkey == flag and (not status):
                    missing = True
                    pos_dup = i
                    # print(pos_dup)
                    break
            if found or missing:
                break
            else:
                new_key = new_key + 1
                bucket = self._hash(new_key)
            
        


        return found, pos_dup, answer, bucket , flag, missing

        
    
    def delete(self, key):
        
        found, pos_dup, answer , bucket, flag,_ = self.search(key)
        if(found):
            bucket[pos_dup] = (flag,"")
        else:
            # print("not found")
            pass
            
        print(self.hashtable)

        return found, answer, bucket
```
