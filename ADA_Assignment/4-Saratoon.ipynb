```python
setup_heap = '''

import random
from random import seed
a = []
seed(18)
for i in range(1000):
    x = random.randint(1,100)
    a.append(x)
# print(a)

def parent(i):
    return i/2

def left(i):
    return 2*i

def right(i):
    return 2*i + 1

def maxheap(a,i):
    l = left(i)
    # print("left is",l)
    r = right(i)
    # print("right is",r)
    if l <= len(a)-1 and a[l] > a[i]:
        largest = l
    else:
        largest = i
    if r <= len(a)-1 and a[r] > a[largest]:
        largest = r
    if largest != i:
        # print(largest)
        a[i],a[largest] = a[largest],a[i]
        maxheap(a,largest)
# maxheap(a,3)

def build_maxheap(a):
    heap_size = len(a) - 1
    for i in range(int((len(a) - 1)/2),0,-1):
        # print("this is i",i)
        maxheap(a,i)
    # print(a)

def heapsort(a):
    # print("start",a)
    a.insert(0,0)
    d = [0] * (len(a)-1)
    build_maxheap(a)
    # print("after build",a)
    for i in range(len(a) - 1,0,-1):
        # print("this is i ",i)
        # print("before swap",a)
        a[1],a[i] = a[i],a[1]
        # print("after swap",a)
        d[i-1] = a[i]
        del a[i]
        maxheap(a,1)
    # print("final result is: ",d)
'''
test_heap = '''
heapsort(a)
'''
```

```python
setup_quick = '''
import random
from random import seed
a = []
seed(18)
for i in range(1000):
    x = random.randint(1,100)
    a.append(x)
# print(a)
def partition(a,p,r):
    x = a[r]
    i = p - 1
    for j in range(p,r):
        if a[j] <= x:
            i = i + 1
            a[i],a[j] = a[j],a[i]
    a[i + 1],a[r] = a[r],a[i + 1]
    return i + 1

def quicksort(a,p,r):
    # print("this is p",p)
    # print("this is r",r)
    if p < r:
        q = partition(a,p,r)
        # print("this is q",q)
        # print(a)
        quicksort(a,p,q-1)
        quicksort(a,q+1,r)
'''
test_quick = '''
quicksort(a[:1000],0,len(a[:1000])-1)
'''
```
